```python
import pandas as pd
```

```python
df=pd.read_csv('Data.csv')
```

```python
df
```

```python
del df['Unnamed: 32']
df
```

```python
X=df.iloc[:,2:]
y=df.iloc[:,1]
```

```python
import seaborn as sns
ax = sns.countplot(df['diagnosis'], label= 'Count')
df['diagnosis'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
y = labelencoder_X_1.fit_transform(y)
```

```python

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python

#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
df.columns
```

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(output_dim=16, init='uniform', activation='relu',input_dim=30))
model.add(Dropout(p=1.0))

#adding the second hidden layer
model.add(Dense(output_dim=16, init='uniform', activation='relu'))
model.add(Dropout(p=1.0))

#adding the output layer
model.add(Dense(output_dim=1, init='uniform', activation='sigmoid'))

```

```python
model.compile(optimizer="Adam", loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, batch_size=100, nb_epoch=150)
```

```python
# Predicting the Test set results
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
sns.heatmap(cm,annot=True)
```
