```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
df=pd.read_csv("Ads.csv")
```

```python
df
```

```python
ads=10
N=10000
total_reward=0
ads_selected=[]

for i in range(0,N):
    ad=random.randrange(ads)
    ads_selected.append(ad)
    reward=df.values[i,ad]
    total_reward+=reward
    



```

```python
plt.hist(ads_selected)
plt.title("Ads")
plt.xlabel("Ads")
plt.ylabel("Frequency of Ads")
plt.show()
```
