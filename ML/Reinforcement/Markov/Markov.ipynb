```python
import pandas as pd
import networkx as nx
```

```python
states=["Sunny","Rainy","Cloudy"]
pi=[0.35,0.35,0.3]
state_space=pd.Series(pi,index=states,name="states")

print(state_space)
print(state_space.sum())
```

```python
df=pd.DataFrame(columns=states,index=states)
df.loc[states[0]]=[0.8,0.05,0.15]
df.loc[states[1]]=[0.2,0.6,0.2]
df.loc[states[2]]=[0.2,0.3,0.5]
df
```

```python
edges={}
def markov_edges(Q):
    for col in Q.columns:
        for idx in Q.index:
            edges[(idx,col)]=Q.loc[idx,col]
    return edges
```

```python
edges=markov_edges(df)
edges
```

```python
G=nx.MultiDiGraph()

G.add_nodes_from(states)

for k,v in edges.items():
    origin,dest=k[0],k[1]
    G.add_edge(origin,dest,label=v,weight=v)

print(G.edges(data=True))
```

```python
prob_1="Sunny"
prob_2="Sunny"
prob_3="Rainy"
rslt=[]
for item in G.edges(data=True):
    p_1=item[0]
    p_2=item[1]
    if (p_1==prob_1 and p_2==prob_2) or (p_1==prob_1 and p_2==prob_3):
        dict=item[2]
        val=dict['label']
        rslt.append(val)

    
print(rslt)
```

```python
import numpy as np
print("probability that tommarrow is sunny and next day is rainy given today is sunny belongs to : {} ".format(np.prod(rslt)))
```
