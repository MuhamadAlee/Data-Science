```python
import pandas as pd
```

```python
df=pd.read_csv("income.csv")
```

```python
df
```

```python
X=df.iloc[:,[1,2]].values
```

```python
from sklearn.cluster import KMeans
errors=[]
for i in range(1,11):
    km=KMeans(n_clusters=i, random_state=0)
    km.fit(X)
    errors.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),errors)
plt.title("Clusters")
plt.xlabel("clusters Numbers")
plt.ylabel("Error")
plt.show()
```

```python
km=KMeans(n_clusters=2,random_state=0)
pred=km.fit_predict(X)
```

```python
pred
```

```python
plt.scatter(X[pred == 0, 0], X[pred == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[pred == 1, 0], X[pred == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()
```
