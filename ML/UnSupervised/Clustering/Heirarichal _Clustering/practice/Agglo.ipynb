```python
import pandas as pd
```

```python
df=pd.read_csv("income.csv")
```

```python
df
```

```python
X=df.iloc[:,[1,2]].values
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
dendo=sch.dendrogram(sch.linkage(X,method="ward"))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
aglo=AgglomerativeClustering(linkage="ward",affinity="euclidean", n_clusters=2)
pred=aglo.fit_predict(X)
```

```python
plt.scatter(X[pred == 0, 0], X[pred == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[pred == 1, 0], X[pred == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.title('Clusters of customers')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()
```
