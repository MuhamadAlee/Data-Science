```python
import pandas as pd
```

```python
df=pd.read_csv("Position_Salaries.csv",usecols=['Level','Salary'])
```

```python
df
```

```python
X=df.iloc[:,:-1].values
Y=df.iloc[:,1].values
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,Y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y,color='red')
plt.plot(X,model.predict(X),color='blue')
```

```python
model.predict([[6.5]])
```

```python
from sklearn.svm import SVR
model=SVR(kernel="rbf")
model.fit(X,Y)
```

```python
model.predict([[6.5]])
```

```python
Y=Y.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
X=sc_x.fit_transform(X)
Y=sc_y.fit_transform(Y)
```

```python
from sklearn.svm import SVR
model=SVR()
model.fit(X,Y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y,color='red')
plt.plot(X,model.predict(X),color='blue')
```

```python

import numpy as np
pred=sc_y.inverse_transform(model.predict(sc_x.transform(np.array([6.5]).reshape(-1,1))))
```
