```python
import pandas as pd
```

```python
df=pd.read_csv("dataset.csv",usecols=['satisfaction_level','average_montly_hours','promotion_last_5years','salary','left'])
```

```python
df
```

```python
dumies=pd.get_dummies(df.salary,prefix="salary")
dumies
```

```python
df=pd.concat([df,dumies],axis='columns')
```

```python
df.drop('salary',axis='columns',inplace=True)
```

```python
df
```

```python
y=df['left']
```

```python
df.drop('left',axis='columns',inplace=True)
```

```python
X=df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
