```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df
```

```python
df.drop(['PassengerId','Name','Ticket','Fare','Cabin','Embarked','SibSp'],axis="columns",inplace=True)
```

```python
df
```

```python
dumies=pd.get_dummies(df.Sex)
```

```python
ndf=pd.concat([df,dumies],axis="columns")
```

```python
ndf
```

```python
ndf.drop(['Sex'],axis="columns",inplace=True)
```

```python
ndf
```

```python
ndf.columns[ndf.isna().any()]
```

```python
ndf.Age = ndf.Age.fillna(ndf.Age.mean())
```

```python
ndf
```

```python
X=ndf.iloc[:,[0,1,2,4,5]]
y=ndf.iloc[:,3]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
