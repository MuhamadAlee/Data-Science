```python
import statistics as st
```

```python
sample=[100,150,200,250,300,100]
```

```python
#mode ( most frquent occurring of an element )

print(st.mode(sample))
```

```python
#median ( centered element in odd list and (elements_length/2 + (elements_length/2)+1 )/2 in case of even list )

print(st.median(sample))
```

```python
#mean ( average of all element ) 

print(st.mean(sample))
```

```python
from sklearn.metrics import explained_variance_score
y_true = [100,150,200,250,300]
y_pred = [130,170,220,260,325]
explained_variance_score(y_true, y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
y_true = [100,150,200,250,300]
y_pred = [130,170,220,260,325]
mean_absolute_error(y_true, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
y_true = [100,150,200,250,300]
y_pred = [130,170,220,260,325]
mean_squared_error(y_true, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
import math as mt
y_true = [100,150,200,250,300]
y_pred = [130,170,220,260,325]
mt.sqrt(mean_squared_error(y_true, y_pred))
```

```python
from sklearn.metrics import mean_squared_log_error
import math as mt
y_true = [100,150,200,250,300]
y_pred = [130,170,220,260,325]
mt.sqrt(mean_squared_log_error(y_true, y_pred))
```
