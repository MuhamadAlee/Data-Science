```python
from sklearn.metrics import r2_score
```

```python
y_true = [21,21,22.8,21.4,18.1,18.7,14.3,24.4,22.8,19.2]
y_pred = [21.5,21.14,26.1,20.2,17.5,19.7,14.9,22.5,25.1,18]
r_sq=r2_score(y_true, y_pred)
```

```python
k=3
ad_r_sq=1-((1-r_sq)*(len(y_pred)-1))/(len(y_pred)-k-1)
```

```python
print("R Square: {}".format(r_sq))
print("Adjusted R Square: {}".format(ad_r_sq))
```
