# fillna

```python
#ffill,bfill
df_fillna=df.fillna(method='ffill')
```

```python
df
```

# Mean|Median|Mode

```python
events=df['event'].fillna(df['event'].mode()[0])
ws=df['windspeed'].fillna(df['windspeed'].mean())
tmp=df['temperature'].fillna(df['temperature'].median())
```

```python
df_mmm=df.copy()
df_mmm['event']=events
df_mmm['temperature']=tmp
df_mmm['windspeed']=ws
```

```python
df_mmm
```
