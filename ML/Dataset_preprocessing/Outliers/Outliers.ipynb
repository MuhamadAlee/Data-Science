# Percentile

```python
import pandas as pd
```

```python
df=pd.read_csv('height.csv')
```

```python
df.head()
```

```python
df=df.drop(['Weight'], axis=1)
```

```python
df.head()
```

```python
df.describe()
```

# Standard Deviation

```python
import pandas as pd
df=pd.read_csv('height.csv')
df.head()
```

```python
df=df.drop(['Weight'],axis=1)
```

```python
df
```

```python
df.describe()
```

```python
max_thresh=df['Height'].mean()+3*df['Height'].std()
min_thresh=df['Height'].mean()-3*df['Height'].std()
```

```python
df=df[(df['Height']<max_thresh)&(df['Height']>min_thresh)]
```

```python
df
```

# Z Score
Zscore = (datapoint - mean)/std


```python
import pandas as pd
df=pd.read_csv('height.csv')
df.head()
```

```python
df=df.drop(['Weight'],axis=1)
```

```python
df
```

```python
 df.describe()
```

```python
df['zscore']=(df['Height']-df['Height'].mean())/df['Height'].std()
```

```python
df=df[df['zscore']<=3]
```

```python
df
```

```python
df=df.drop(['zscore'],axis=1)
```

```python
df
```

# IQR
IQR=Q3-Q1

max=q3+IQR*1.5


min=q1-IQR*1.5

```python
import pandas as pd
df=pd.read_csv('height.csv')
df.head()
```

```python
df=df.drop(['Weight'],axis=1)
```

```python
df
```

```python
IQR=df['Height'].quantile(0.75)-df['Height'].quantile(0.25)
```

```python
max_thresh=df['Height'].quantile(0.75)+1.5*IQR
min_thresh=df['Height'].quantile(0.25)-1.5*IQR
```

```python
df=df[(df['Height']<max_thresh)&(df['Height']>min_thresh)]
```

```python
df
```
